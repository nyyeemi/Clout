```python
import random
print(random.gauss(33.93, 2.198))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

n = 1000
l = np.random.normal(loc=33.93, scale=2.198, size=n)

# Plot histogram density
count, bins, _ = plt.hist(l, bins=30, density=True, alpha=0.6, edgecolor='black')

# Overlay the theoretical PDF
mu, sigma = 33.93, 2.198
x = np.linspace(min(l), max(l), 100)
plt.plot(x, norm.pdf(x, mu, sigma), 'r-', lw=2)

plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Distribution with Normal PDF")
plt.show()
```
